Coeficiente de determinación general

```python
# Crear un DataFrame temporal para la limpieza
df_for_split = icfes_sin_nan.copy()

# Eliminar 'PERCENTIL_ESPECIAL_GLOBAL' debido a la gran cantidad de NaNs
if 'PERCENTIL_ESPECIAL_GLOBAL' in df_for_split.columns:
    df_for_split = df_for_split.drop(columns=['PERCENTIL_ESPECIAL_GLOBAL'])

# Identificar columnas numéricas para el modelo, incluyendo la variable objetivo
numeric_cols_for_model = df_for_split.select_dtypes(include=np.number).columns

# Filtrar el DataFrame para incluir solo columnas numéricas
df_for_split_numeric = df_for_split[numeric_cols_for_model]

# Eliminar filas con cualquier valor NaN restante del DataFrame completamente numérico
df_for_split_numeric_clean = df_for_split_numeric.dropna()

# Ahora definir X e Y a partir del DataFrame completamente limpio y numérico
# Y es ESTU_INSE_INDIVIDUAL, X son todas las demás columnas numéricas
X = df_for_split_numeric_clean.drop("ESTU_INSE_INDIVIDUAL", axis=1)
Y = df_for_split_numeric_clean["ESTU_INSE_INDIVIDUAL"]

# Realizar la división de entrenamiento-prueba
trainX, testX, trainY, testY = train_test_split(X, Y, test_size = 0.2, random_state = 0)
trainX.head(4)
```

```python
modelo_lineal = LinearRegression()
regresion = modelo_lineal.fit(trainX, trainY)
pred_lineal = regresion.predict(testX)
```

```python
r2_lineal = regresion.score(testX, testY)
r2_lineal
#Coeficiente de determinación = 0.8115149039204218
```

Diagrama con el plano de dispersión

```python
XY, Z = icfes_sin_nan[["PUNT_LECTURA_CRITICA", "PUNT_INGLES"]], icfes_sin_nan["ESTU_INSE_INDIVIDUAL"]
trainXY, testXY, trainZ, testZ = train_test_split(XY, Z, test_size = 0.2, random_state =0)
trainXY.head(4)
modelo1 = LinearRegression()
modelo1.fit(trainXY, trainZ)
modelo1.score(testXY, testZ)
```

```python
array = modelo1.coef_
array
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(projection="3d")
xx, yy = np.meshgrid(np.linspace(35, 85, num=10), np.linspace(35, 95, num=10))
plano_reg = array[0]*xx + array[1]*yy + modelo1.intercept_
ax.plot_surface(xx, yy, plano_reg, alpha=0.1, cmap="plasma")

sc=ax.scatter(trainXY["PUNT_LECTURA_CRITICA"], trainXY["PUNT_INGLES"], trainZ, c=trainZ, cmap="plasma", s=20, edgecolors='none', alpha=0.6)
ax.scatter(testXY["PUNT_LECTURA_CRITICA"], testXY["PUNT_INGLES"], testZ, c=testZ, cmap="plasma", s=20, edgecolors='none', alpha=0.6)


ax.set_xlabel("Puntaje lectura critica")
ax.set_ylabel("Puntaje ingles")
ax.set_zlabel("Indice socio-economico")
ax.set_title("Regresión lineal con multiples variables")

cbar = plt.colorbar(sc, ax=ax, shrink=0.45, pad=0.1)
cbar.set_label('INSE Individual', fontsize=9)

plt.show
```
