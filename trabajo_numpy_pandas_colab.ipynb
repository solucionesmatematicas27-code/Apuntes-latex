# **2. Numpy y Pandas sus Utilidades.**

## ¿Qué es NumPy?

NumPy es una librería de Python orientada al cálculo numérico y al manejo eficiente de arreglos multidimensionales llamados `ndarray`. El cálculo  orientado a `array` en Phyton su origen fue 1995 cuando **Jim Huguinin** creó la libreria **Númeric**.

## ¿Qué es Pandas?

Pandas es una librería de Python especializada en análisis y manipulación de datos estructurados.



| Aspecto | NumPy | Pandas |
|---|---|---|
| Uso principal | Computación numérica | Análisis y manipulación de datos |
| Operaciones matemáticas | Operaciones matemáticas rápidas | Estadística descriptiva |
| Manejo de datos | Vectores y matrices | Tablas y DataFrames |
| Álgebra lineal | Sí | Limitado |
| Estadística | Sí | Sí |
| Números aleatorios | Generación de números aleatorios | No es su función principal |
| Rendimiento | Procesamiento eficiente de grandes volúmenes de datos | Muy eficiente para análisis tabular |
| Lectura de archivos | Básica | Lectura de CSV y Excel |
| Limpieza de datos | Limitada | Excelente |
| Filtrado y agrupación | Básico | Avanzado |
| Estructura principal | `ndarray` | `Series` y `DataFrame` |
| Aplicaciones comunes | Matemáticas, IA, simulaciones | Ciencia de datos, análisis y reportes |

---

## Diferencias entre NumPy y Pandas


| NumPy | Pandas |
|---|---|
| Trabaja con arrays | Trabaja con tablas |
| Más rápido en cálculos numéricos | Más práctico para análisis de datos |
| Usa ndarray | Usa Series y DataFrame |
| Orientado a matemática | Orientado a análisis de información |

# **5. Ejemplos de arreglos con NumPy**

Mostrando algunos ejemplos donde se utlizará las respectivas librerias en diferentes contextos.


```python

import numpy as np

# Temperaturas semanales
temperaturas = np.array([22, 24, 21, 23, 25, 26, 22])

print("Temperaturas:", temperaturas)

print("Promedio:", np.mean(temperaturas))
print("Temperatura máxima:", np.max(temperaturas))
print("Temperatura mínima:", np.min(temperaturas))
print("Desviación estándar:", np.std(temperaturas))
```


## Operaciones vectorizadas con NumPy


```python

ventas = np.array([100, 120, 90, 150])

# Incremento del 10%
ventas_actualizadas = ventas * 1.10

print("Ventas originales:", ventas)
print("Ventas actualizadas:", ventas_actualizadas)
```


# **6. Ejemplos de DataFrame con Pandas**


```python

import pandas as pd

datos = {
    "Producto": ["Laptop", "Mouse", "Teclado", "Monitor"],
    "Precio": [2500, 80, 150, 900],
    "Cantidad": [5, 20, 15, 8]
}

df = pd.DataFrame(datos)

print(df)
```


## Filtrar productos con precio mayor a 500


```python

filtro = df[df["Precio"] > 500]

print(filtro)
```


## Crear nueva columna de valor total


```python

df["Valor_Total"] = df["Precio"] * df["Cantidad"]

print(df)
```


## Estadísticas del DataFrame


```python

print(df.describe())
```
